```python
ks = pd.concat([failed,success ])
ks
```

```python
main = pd.get_dummies(ks['main_category'], drop_first=True)
main

```

```python
state = pd.get_dummies(ks['state'], drop_first=True)
state
```

```python
test = pd.concat([ks,state,main], axis=1)
test
```

```python
import datetime
```

```python
test[['deadline', 'launched']] = test[['deadline', 'launched']].apply(pd.to_datetime, errors='coerce', infer_datetime_format=True)
```

```python
test['length'] = test['deadline'] - test['launched']
```

```python
def dayFormatter(timedelta):
    numdays = timedelta.days
    return numdays
test['length'].apply(lambda x: dayFormatter(x))
```

```python
features = test.drop(['ID', 'name', 'category', 'main_category', 'currency', 'deadline', 'launched','state','country', 'usd pledged','usd_pledged_real','usd_goal_real', 'length'], axis=1)
```

```python
features.info()
```

```python
yval = ['successful']
yval
```

```python
#Importing train/test split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(features.drop(['successful'],axis=1), 
                                                    features['successful'], test_size=0.10, 
                                                    random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
#Initializing the logistic regressor
logmodel = LogisticRegression(max_iter=1000)
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```
